```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

intervalos = list(range(15, 65, 5))

def graficos_continua(df, columna):
    data = df[columna].dropna()

    plt.figure(figsize=(8,5))
    # Histograma con seaborn (solo para estilo)
    ax = sns.histplot(data,bins=intervalos,color="#2a9d8f",edgecolor="white")

    # Calcular conteos por intervalo para colocar etiquetas
    counts, bin_edges = np.histogram(data, bins=intervalos)
    for count, x_left, x_right in zip(counts, bin_edges[:-1], bin_edges[1:]):
        ax.text(
            (x_left + x_right) / 2,
            count + 0.5,
            str(int(count)),
            ha='center',
            va='bottom',
            fontsize=10,
            weight='bold'
        )

    plt.title(f"Distribución de {columna}", fontsize=14, weight="bold")
    plt.xlabel(columna, fontsize=12)
    plt.ylabel("Frecuencia", fontsize=12)
    plt.xticks(intervalos)
    plt.tight_layout()
    plt.show()

# Uso
graficos_continua(df, "EDAD")
```

```python
def grafico_barras(df, columna, color="#2a9d8f"):
    # Contar y ordenar categorías
    counts = df[columna].value_counts()
    
    plt.figure(figsize=(9,4))
    ax = sns.barplot(x=counts.index,y=counts.values,color=color,edgecolor="white")

    # Etiquetas de frecuencia sobre cada barra
    for p in ax.patches:
        ax.annotate(f"{int(p.get_height())}",
                    (p.get_x() + p.get_width()/2, p.get_height()),
                    ha='center', va='bottom', fontsize=10, weight='bold')

    plt.title(f"Distribución de {columna}", fontsize=14, weight="bold")
    plt.xlabel(columna, fontsize=12)
    plt.ylabel("Frecuencia", fontsize=12)
    plt.grid(axis="y", linestyle="--", alpha=0.4)
    plt.tight_layout()
    plt.show()


# Duración total en semestres
grafico_barras(df, "DURACION TOTAL CARRERA (SEMESTRES)")
```

```python

def graficos_discreta(df, columna, ultimos=5, color="#2a9d8f"):
    # Asegurar que sea numérico
    df[columna] = pd.to_numeric(df[columna], errors="coerce")
    
    # Filtrar últimos 'ultimos' años
    max_anio = int(df[columna].max())
    filtro = df[df[columna] >= max_anio - (ultimos - 1)]
    
    # Orden ascendente
    orden = sorted(filtro[columna].dropna().unique())
    
    plt.figure(figsize=(7,4))
    ax = sns.countplot(data=filtro,x=columna,order=orden,color=color,edgecolor="white")
    
    # Etiquetas de frecuencia en cada barra
    for p in ax.patches:
        ax.annotate(
            f"{int(p.get_height())}",
            (p.get_x() + p.get_width()/2, p.get_height()),
            ha="center", va="bottom",
            fontsize=10, weight="bold"
        )
    
    plt.title(f"Frecuencia de {columna} (últimos {ultimos} años)", fontsize=14, weight="bold")
    plt.xlabel(columna, fontsize=12)
    plt.ylabel("Frecuencia", fontsize=12)
    plt.grid(axis="y", linestyle="--", alpha=0.4)
    plt.tight_layout()
    plt.show()

# Ejemplo de uso
graficos_discreta(df, "AÑO INGRESO", ultimos=10)
```
